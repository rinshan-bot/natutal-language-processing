```python
import numpy as np
import matplotlib.pyplot as plt

def mean_error(act, pred):
    diff = pred - act
    return diff


def mean_squared_error(act, pred):
    diff = pred - act
    differences_squared = diff ** 2
    return differences_squared

act = np.array([1.1,2,1.7])
pred = np.array([1,1.7,1.5])

pred_one = mean_error(act,pred)
pred_two = mean_squared_error(act,pred)
```

```python
plt.figure(figsize=(10,5))
# plt.plot(pred_one)
plt.plot(pred_two)
plt.show()
```
